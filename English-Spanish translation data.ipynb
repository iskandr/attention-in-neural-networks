```python
from normalization import tokenize
```

```python
spanish_lines = []
with open("europarl-v7.es-en.es", "r") as f:
    for line in f:
        spanish_lines.append(tokenize(line))
print("Read %d lines of Spanish" % len(spanish_lines))
english_lines = []
with open("europarl-v7.es-en.en", "r") as f:
    for line in f:
        english_lines.append(tokenize(line))
        
print("Read %d lines of English" % len(english_lines))
assert len(spanish_lines) == len(english_lines)
```

```python
english_lines[0:10], spanish_lines[0:10]
```

```python
longest_english_line = max([len(line) for line in english_lines])
print("Longest English line: %d" % longest_english_line)
longest_spanish_line = max([len(line) for line in spanish_lines])
print("Longest Spanish line: %d" % longest_spanish_line)
```

```python
for MAX_SENTENCE_LENGTH in [60, 70, 80, 90]:
    print("MAX LENGTH = %d" % MAX_SENTENCE_LENGTH)
    number_long_lines_english = sum([len(line) > MAX_SENTENCE_LENGTH for line in english_lines])
    print("# English lines > 100 words: %d/%d (%0.3f%%)" % (
        number_long_lines_english, 
        len(english_lines),
        100 * number_long_lines_english / len(english_lines)))

    number_long_lines_spanish = sum([len(line) > MAX_SENTENCE_LENGTH for line in spanish_lines])
    print("# Spanish lines > 100 words: %d/%d (%0.3f%%)" % (
        number_long_lines_spanish, 
        len(spanish_lines),
        100 * number_long_lines_spanish / len(spanish_lines)))
```

```python
MAX_SENTENCE_LENGTH = 90

unique_spanish_words = set([])
unique_english_words = set([])
short_spanish_lines = []
short_english_lines = []
for english_line, spanish_line in zip(english_lines, spanish_lines):
    if len(english_line) < MAX_SENTENCE_LENGTH and len(spanish_line) < MAX_SENTENCE_LENGTH:
        unique_spanish_words.update(spanish_line)
        unique_english_words.update(english_line)
        short_english_lines.append(english_line)
        short_spanish_lines.append(spanish_line)

print("%d unique Spanish words" % len(unique_spanish_words))
print("%d unique English words" % len(unique_english_words))
```

```python
combined_words = unique_english_words.union(unique_spanish_words)
print("%d words combined" % len(combined_words))
```

```python
for i, w in enumerate(combined_words):
    print("%d) %s" % (i + 1, w))
    if i > 1000:
        break
```

```python

# give every word a base-1 index, allowing us to use 0 for null padding
word2index = {
    word: index + 1
    for index, word
    in enumerate(combined_words)
}
index2word = {
    index: word
    for (word, index)
    in word2index.items()
}
```

```python
n_lines = len(short_english_lines)
assert len(short_spanish_lines) == n_lines
```

```python
shape = (n_lines, MAX_SENTENCE_LENGTH)
X_english = np.zeros(shape, dtype="int32")
X_spanish = np.zeros(shape, dtype="int32")
```

```python
for (language, language_lines, X) in [("english", short_english_lines, X_english), ("spanish", short_spanish_lines, X_spanish)]:
    print("Processing %s" % language)
    for i, line in enumerate(language_lines):
        for j, word in enumerate(line):
            X[i, j] = word2index[word]
```

```python
data_to_save = {
    "X_english": X_english,
    "X_spanish": X_spanish,
    "line_lengths_english": np.array([len(line) for line in short_english_lines]),
    "line_lengths_spanish": np.array([len(line) for line in short_spanish_lines]),
    "word2index": word2index,
    "index2word": index2word,
    "max_length": MAX_SENTENCE_LENGTH,
}
```

```python
import pickle
with open("encoded_corpus.pickle", "wb") as f:
    pickle.dump(data_to_save, f)
```
